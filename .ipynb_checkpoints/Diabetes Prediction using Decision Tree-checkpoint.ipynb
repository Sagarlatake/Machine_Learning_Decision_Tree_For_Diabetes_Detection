# Diabetes Prediction using Decision Tree

```python
# import libraries

import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
```

```python
# load dataset

data = pd.read_csv("dataset.csv")
data.head()
```

```python
# Lets understand the data
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
# import encoder

from sklearn.preprocessing import LabelEncoder
```

```python
# create a alias for label encoder

le = LabelEncoder()
```

```python
# lets take all columns in one list

cols = data.columns
cols
```

```python
# we have dropped Age column as it is already a numeric one
cols_1 = cols.drop('Age')
cols_1
```

```python
data_1 = data[cols_1]
data_1.head()
```

```python
# apply label encoding on data_1
# as label encoding is to applied on one column at one time
for colmn in cols_1:
    data_1[colmn] = le.fit_transform(data_1[colmn])
```

```python
# check encoded data
data_1.head()
```

```python
# add age column to data_1

data_1['Age'] = data['Age']
```

```python
data_1.head()
```

```python
data_1.describe()
```

```python
data_1.info()
```

```python
Y = pd.DataFrame()
Y['class']= data_1['class']
```

```python
c = data_1.columns
```

```python
c_1 = c.drop('class')
```

```python
X = data_1[c_1]
```

```python
X.head()
```

```python
Y.head()
```

```python
# Lets use Standard scaler to scale values

from sklearn.preprocessing import MinMaxScaler

sc = MinMaxScaler()
```

```python
# Use sc to scale features

x_scaled = sc.fit_transform(X)
x_scaled = pd.DataFrame(x_scaled)
```

```python
# Lets split train and test data 

from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest = train_test_split(x_scaled,Y)
```

```python
x_scaled.head()
```

```python
xtrain.shape
```

```python
ytrain.shape
```

```python
# Now Create a model using scaled features

dt = DecisionTreeClassifier()
    
```

```python
dt.fit(x_scaled,Y)
```

```python
dt.predict(x_scaled)
```

```python
dt.score(x_scaled,Y)
```

```python
dt1 = DecisionTreeClassifier()
dt1.fit(xtrain,ytrain)
```

```python
pred = dt1.predict(xtest)
pred
```

```python
dt1.score(xtest,ytest)
```

```python
# Lets do the analysis of results
from sklearn.metrics import accuracy_score,auc,classification_report,confusion_matrix

ac = accuracy_score(pred,ytest)

cr = classification_report(pred,ytest)
cm = confusion_matrix(pred,ytest)
```

```python
print("Accuracy on test Data:")
print(ac)
print("confusion matrix on test Data:")
print(cm)
print("Clasification report on test Data:")
print(cr)
```
